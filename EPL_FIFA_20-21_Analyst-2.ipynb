```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("C:/Code/Pandas from VScode/Kaggle/FIFA/EPL_20_21.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
# creat 2 new columns

df["Mints_per_match"] = (df["Mins"]/df["Matches"]).astype(int)
df["Goals_per_match"] = df["Goals"]/df["Matches"].astype(float)
df.head()
```

```python
# total Goals
total_goals = df["Goals"].sum()
total_goals
```

```python
# total penalty Goals
total_penlt_goals = df["Penalty_Goals"].sum()
total_penlt_goals
```

```python
# total penalty attempts
total_penlt_attempts = df["Penalty_Attempted"].sum()
total_penlt_attempts
```

```python
# pie chart between penalty Missed vs Scored 

plt.figure(figsize=(13,6))
penlt_not_scored = df["Penalty_Attempted"].sum() - total_penlt_goals
data = [penlt_not_scored, total_penlt_goals]
label = ['Penalty Missed','Penalty Scored']
plt.figure(facecolor='white')
plt.title("Penalty Missed Vs Scored",fontsize=20)
color = sns.color_palette('Set2')
plt.pie(data , labels= label, colors=color , autopct='%0.0f%%')
plt.show
```

```python
# Unique position 
unique_value = df["Position"].unique()
unique_value
```

```python
# total FW players
total_forward_player = df[df["Position"]== "FW"]
total_forward_player
```

```python
# player from diffrent nation
np.size(df["Nationality"].unique())
```

```python
# most player from which country "bar chart"

nationality = df.groupby("Nationality").size().sort_values(ascending=False)
nationality.head(10).plot(kind = 'bar', figsize=(12,6), color=sns.color_palette("magma"))
```

```python
# club with maximum player in thier squard

df["Club"].value_counts().nlargest(5).plot(kind= 'bar', color=sns.color_palette("viridis"))
```

```python
# club with least player in thier squard

df["Club"].value_counts().nsmallest(5).plot(kind= 'bar', color=sns.color_palette("viridis"))
```

```python
# player based on age group

Under20 = df[df["Age"] <=20]
age20_25 = df[(df["Age"] >20) & (df["Age"] <=25)]
age25_30 = df[(df["Age"] >25) & (df["Age"] <=30)]
above30 = df[df["Age"] >30]
```

```python
x = np.array([Under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above30['Name'].count()])
mylabels = ["<=20", ">20 & <=25" ,">25 & <=30", ">30"]
plt.figure(facecolor='white') #background has white
plt.title("Total players with age Groups", fontsize=20)
plt.pie(x, labels= mylabels, autopct="%.1f%%")
plt.show()
```

```python
# total under 20 player in each club
player_under_20 = df[df["Age"] <20]
player_under_20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette("cubehelix"))
```

```python
# under 20 age players in Manchester United
player_under_20[player_under_20["Club"]== "Manchester United"]
```

```python
# under 20 age players in Chelsea
player_under_20[player_under_20["Club"]== "Chelsea"]
```

```python
# avaerage age of players in each club in boxplot

plt.figure(figsize=(12, 6))
sns.boxplot(x = 'Club', y= 'Age', data=df)
plt.xticks(rotation = 90)
plt.legend()
```

```python
num_player = df.groupby('Club').size()
data = df.groupby('Club')['Age'].sum()/num_player
data.sort_values(ascending=False)
```

```python
# total assistant from each club

assits_by_club = pd.DataFrame(df.groupby('Club', as_index=False)["Assists"].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax = sns.barplot(x='Club', y='Assists',data=assits_by_club.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=30)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Club vs Total Assists',fontsize=20)
```

```python
# top 10 Assists

top_10_assists = df[['Name', 'Club', 'Assists','Matches']].nlargest(n=10, columns = "Assists")
top_10_assists
```

```python
goals_by_club = pd.DataFrame(df.groupby('Club', as_index=False)["Goals"].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax = sns.barplot(x='Club', y='Goals',data=goals_by_club.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Club vs Total Goals',fontsize=20)
```

```python
# most goal by players
top_10_goals = df[['Name', 'Club','Goals','Matches']].nlargest(n=10, columns = "Goals")
top_10_goals
```

```python
# goals per match
top_10_goals_permatch = df[['Name','Goals_per_match','Matches','Goals']].nlargest(n=10, columns="Goals_per_match")
top_10_goals_permatch
```

```python
# pie chart - goals with assists and without assists

plt.figure(figsize=(14,7))
assists = df['Assists'].sum()
data = [total_goals - assists, assists]
label = ['Goals with Assists','Goals without Assists']
color = sns.color_palette('Set1')
plt.title("Goals with Assists and Without Assists")
plt.pie(data, labels=label, colors=color, autopct='%.0f%%')
plt.show()
```

```python
# top 10 players with most yellow cards

most_yellow = df.sort_values(by="Yellow_Cards",ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Players with most number of Yellow card")
c = sns.barplot(x=most_yellow["Name"],y=most_yellow["Yellow_Cards"], label= "Players", color='yellow')
plt.ylabel('Number of yellow card')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
```
